```python
import os
import json
from PIL import Image
import matplotlib.pyplot as plt

def visualize_images_and_json(folder_path):
    #     
    files = os.listdir(folder_path)

    # JPG  JSON  
    jpg_files = [file for file in files if file.lower().endswith('.jpg') or file.lower().endswith('.jpeg')]
    json_files = [file for file in files if file.lower().endswith('.json')]

    #   3  (  3)
    jpg_files_to_visualize = jpg_files[:3]

    for jpg_file in jpg_files_to_visualize:
        #   
        jpg_path = os.path.join(folder_path, jpg_file)

        #  
        image = Image.open(jpg_path)
        plt.imshow(image)
        plt.title(jpg_file)  #    
        plt.axis('off')
        plt.show()

        #     JSON  
        json_file_name = os.path.splitext(jpg_file)[0] + '.json'
        if json_file_name in json_files:
            # JSON  
            json_path = os.path.join(folder_path, json_file_name)

            # JSON   
            with open(json_path, 'r') as f:
                json_data = json.load(f)
                print(f"Contents of {json_file_name}:")
                print(json_data)
        else:
            print(f"No JSON file found for {jpg_file}")

#    
folder_path = r'C:\Users\tata\Desktop\ \\01.\1.Training\\TS_2.1\TS_2.1'

#  JSON    
visualize_images_and_json(folder_path)
```
