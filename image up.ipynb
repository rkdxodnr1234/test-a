```python
import os
import json
from PIL import Image
import matplotlib.pyplot as plt

def visualize_images_and_json(folder_path):
    # 폴더 내의 파일 목록 확인
    files = os.listdir(folder_path)

    # JPG 파일과 JSON 파일 구분
    jpg_files = [file for file in files if file.lower().endswith('.jpg') or file.lower().endswith('.jpeg')]
    json_files = [file for file in files if file.lower().endswith('.json')]

    # 이미지 파일 3개 선택 (첫 번째부터 3개)
    jpg_files_to_visualize = jpg_files[:3]

    for jpg_file in jpg_files_to_visualize:
        # 이미지 파일 경로
        jpg_path = os.path.join(folder_path, jpg_file)

        # 이미지 표시
        image = Image.open(jpg_path)
        plt.imshow(image)
        plt.title(jpg_file)  # 이미지 파일명을 제목으로 설정
        plt.axis('off')
        plt.show()

        # 해당 이미지와 같은 이름의 JSON 파일 찾기
        json_file_name = os.path.splitext(jpg_file)[0] + '.json'
        if json_file_name in json_files:
            # JSON 파일 경로
            json_path = os.path.join(folder_path, json_file_name)

            # JSON 파일 내용 출력
            with open(json_path, 'r') as f:
                json_data = json.load(f)
                print(f"Contents of {json_file_name}:")
                print(json_data)
        else:
            print(f"No JSON file found for {jpg_file}")

# 검색할 폴더 경로 설정
folder_path = r'C:\Users\tata\Desktop\딸기 데이터\설향\01.데이터\1.Training\원천데이터\TS_2.설향1\TS_2.설향1'

# 이미지와 JSON 파일 시각화 및 출력
visualize_images_and_json(folder_path)
```

```python
# 이미지 파일이 있는 폴더 경로
image_folder = 'C:/Users/tata/Desktop/딸기 데이터/설향/01.데이터/1.Training/extracted_images'

# 이미지 데이터와 라벨을 저장할 리스트
images = []
labels = []

# 이미지 파일과 라벨 정보 매칭
for root, dirs, files in os.walk(image_folder):
    for file in files:
        if file.endswith('.jpg'):
            image_path = os.path.join(root, file)
            img = Image.open(image_path)
            img = img.resize((224, 224))  # 이미지 크기 조정
            img_array = img_to_array(img)  # 이미지를 배열로 변환
            img_array = img_array / 255.0  # 이미지 스케일링 (0 ~ 1 사이 값으로 변환)
            images.append(img_array)

            # 이미지 파일 이름에서 라벨 정보 추출 (예: 이미지 파일 이름이 'image001.jpg'라면 'image001' 부분을 추출)
            image_filename = os.path.splitext(file)[0]
            if image_filename in labels_dict['growth_stage']:
                label = labels_dict['growth_stage'][image_filename]  # 라벨 딕셔너리에서 해당 이미지의 growth_stage 가져오기
                labels.append(label)
            else:
                labels.append('unknown')  # 라벨이 없는 경우 처리 (예: 라벨 딕셔너리에 해당 이미지 이름이 없는 경우)

# NumPy 배열로 변환
images = np.array(images)
labels = np.array(labels)

# 데이터셋 분할 (훈련 데이터와 검증 데이터)
X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.2, random_state=42)

print("Dataset prepared successfully.")
```
